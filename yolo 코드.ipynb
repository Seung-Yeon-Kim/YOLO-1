### dakrnet을 다운로드 받은후 <br> GPU=0, CUDNN=0, OPENCV=0 → GPU=1, CUDNN=1, OPENCV=1 수정, <br> 7행에 ARCH 전부 주석처리, 14행<br> `# ARCH= -gencode arch=compute_52,code=compute_52`
이 코드를 `ARCH= -gencode arch=compute_75,code=compute_75` 로 수정, 50, 52행에 cuda뒤에 CUDA버젼 확인후 맞게끔 써주면 된다. 여기 코랩에서는 10.0 및 10.1 있어서 cuda-10.0혹은 10.1로 사용한다.
